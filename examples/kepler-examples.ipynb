# Étape 1 : visualisation simple des flux 1789

J'ai joué à partir de la doc de kepler : https://docs.kepler.gl/docs/keplergl-jupyter#demo-notebooks

```python
# Load an empty map
from keplergl import KeplerGl
from poitousprint import Portic
import pandas as pd

portic_client = Portic()
```

```python
travels = portic_client.get_flows(year=1789)
```

```python
# setting kepler config
config = {
    'version': 'v1',
    'config': {
        # centering the map on the region
        'mapState': {
            'latitude': 45.6876849,
            'longitude': -1.15,
            'zoom': 5.2
        }
    }
}
# fixing data types issues
travels_clean = []
for travel in travels:
    final_travel = travel.copy()
    COORDS_KEYS = ["departure_latitude", "departure_longitude", "destination_latitude", "destination_longitude"]
    for key in COORDS_KEYS:
        final_travel[key] = float(final_travel[key])
    travels_clean.append(final_travel)
# creating a dataframe with all the travels
df = pd.DataFrame(travels_clean)

flux_map = KeplerGl(config=config, data={'data_1': df})
flux_map
```

# Étape 2 : filtrer sur les flux du datasprint

Je filtre les flux à source_subset = Poitou_1789

```python
travels_clean = []
for travel in travels:
    final_travel = travel.copy()
    if final_travel["source_subset"] == "Poitou_1789":
        COORDS_KEYS = ["departure_latitude", "departure_longitude", "destination_latitude", "destination_longitude"]
        for key in COORDS_KEYS:
            final_travel[key] = float(final_travel[key])
        travels_clean.append(final_travel)
# creating a dataframe with all the travels
df = pd.DataFrame(travels_clean)

flux_map = KeplerGl(config=config, height=800, data={'data_1': df})
flux_map
```
