## Objectif : 
Pour les produit Navigo faisant partie des produits qu'on retrouve à la fois dans Toflit18 et Portic pour le sprint (avec le croisement selon la classification "Revolution and Empire")
=> Requête de Silvia : savoir à quoi ces produits Navigo correspondent au niveau source


```python
from poitousprint import Portic, Toflit, get_pointcalls_commodity_purposes_as_toflit_product

toflit_client = Toflit()
portic_client = Portic()
```

```python
# choix d'une classification d'alignement
chosen_classification = 'product_revolutionempire'
```

```python
# stockage dans une liste des produits communs Navigo / Portic pour datasprint avec cette classification

common_products_revempire_datasprint = [
'Fruits secs et noix',
'Graines diverses et mélangées',
'Fromage',
'Brai, goudron et résines',
'Bière, cidre et vinaigre',
'Matières premières diverses',
'Légumes',
'Cordage et filets',
'Sel',
'Beurre',
'Ouvrages de bois, liège et jonc',
'Savon',
'Ouvrages divers et mélangés',
'Ouvrages métalliques',
'Fer',
'Lin',
'Blé et autres grains comestibles',
'Graines grasses',
'Bâtiments de mer et apparaux',
'Chairs et œufs',
'Farine, gruau, biscuits et pâtes',
'Habits et lingerie',
'Divers incertains',
'Chanvre',
'Porcelaine, poterie, faïence et ouvrages de terre',
'Graisses diverses',
'Verrerie',
'Eaux-de-vie et liqueurs',
'Divers mélangés',
'Animaux',
'Sucre',
'Café',
'Produits inconnus',
'Vins divers',
'Fruits frais',
'Bois divers hors combustible',
'Cuirs, peaux et pelleterie',
'Huiles diverses',
'Meubles, glaces et miroirs',
'Combustibles dont charbons et bois',
'Subsistances diverses et mélangées',
'Coton non transformé',
'Laines non transformées diverses',
'Armes, munitions, poudre et pierres à fusil',
'Pêche et fruits de mer',
'Cuivre'
]
```

```python
# récupération des pointcalls Portic qui concernent le datasprint
pointcalls_datasprint = portic_client.get_pointcalls(
 year = 1789,
 pointcall_admiralty = ['La Rochelle', 'Marennes', 'Sables d\'Olonne']
)

# enrichissement de ces pointcalls avec la propriété 'commodity_as_toflit' qui nous donne l'équivalent des produits navigo dans Toflit18 avec la classification choisie
croisement_produits = get_pointcalls_commodity_purposes_as_toflit_product(pointcalls_datasprint, product_classification=chosen_classification)

croisement_produits[0:10]
```

```python
navigo_products_for_datasprint = {}

for pointcall in croisement_produits:
    if pointcall['commodity_purposes'] is not None:
        for commodity in pointcall['commodity_purposes']:
            # objectif de format
            """
            {
            'prodrevempire':list(matches),
            ...
            }
            """
            
            # si nouveau prod rev & empire qui fait bien partie des produits communs on ajoute juste ce produit comme clé du dict et on commence à remplir le set de produits navigo
            if commodity['commodity_as_toflit'] not in navigo_products_for_datasprint.keys() and commodity['commodity_as_toflit'] in common_products_revempire_datasprint:
                navigo_products_for_datasprint[commodity['commodity_as_toflit']] = [commodity['commodity_purpose']]
                    
            # si prod rev empire pas nouveau mais qui fait bien partie des produits communs, on ajoute juste product source navigo dans list de matches
            elif commodity['commodity_as_toflit'] in navigo_products_for_datasprint.keys() and commodity['commodity_as_toflit'] in common_products_revempire_datasprint: 
                # on ne fait l'ajout que si le produit source ne fait pas déjà partie de la liste de matches
               if commodity['commodity_purpose'] not in navigo_products_for_datasprint[commodity['commodity_as_toflit']]:
                    navigo_products_for_datasprint[commodity['commodity_as_toflit']].append(commodity['commodity_purpose'])


navigo_products_for_datasprint
```

```python
# écriture dans un fichier csv
import csv
from csv import DictWriter

# tableau qui contient une ligne par produit source navigo 
with open('dumps/produits_source_navigo_dans_communs_datasprint_revempire_pour_silvia.csv', 'w', newline='') as csvfile:
        fieldnames = ['catégorie_toflit_revolutionempire','produit_source_navigo']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for cat_toflit_revempire, matchs_produits_sources_navigo in navigo_products_for_datasprint.items():
            k=0
            
            for produit_source_navigo in matchs_produits_sources_navigo:
                if k == 0:
                    writer.writerow({
                        'catégorie_toflit_revolutionempire':cat_toflit_revempire,
                        'produit_source_navigo':produit_source_navigo
                    })
                    k += 1
                else:
                    writer.writerow({
                        'produit_source_navigo':produit_source_navigo
                    })
```
